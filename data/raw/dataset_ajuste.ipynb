```python
import pandas as pd
import numpy as np
import anthropic
import os

api_key = os.getenv("CLAUDE_KEY")
if not api_key:
    print("AVISO: Variável de ambiente ANTHROPIC_API_KEY não encontrada!")
    print("Defina sua chave API do Anthropic na variável de ambiente.")

anthropic.api_key = api_key

MODEL = "claude-3-5-haiku-20241022"
```

```python
df_prompts = pd.read_csv("portugues_poems_prompts_generator.csv")


df_portuguese_poems = pd.read_csv("portuguese-poems.csv")
```

```python
client = anthropic.Anthropic(api_key=api_key)
```

```python

def gerar_poema(prompt):
    """
    Gera um poema usando a API da Anthropic
    """
    if not client:
        return "API key não configurada"
    
    try:
        response = client.messages.create(
            model="claude-3-5-haiku-20241022",
            max_tokens=1000,
            temperature=0.8,
            messages=[
                {
                    "role": "user", 
                    "content": prompt + "Algumas regras: O resultado deve conter apenas o poema, sem nenhum outro texto. O poema deve ser em português brasileiro. O poema deve ser em formato de poema, não deve conter nenhum outro formato."
                }
            ]
        )
        return response.content[0].text
    except Exception as e:
        return f"Erro ao gerar poema: {str(e)}"

df_prompts["poem_generator"] = df_prompts["final prompt"].apply(gerar_poema)

df_prompts.head(10)
```
